```python
lista=[12,-2,4,8,29,45,78,36,-17,2,12,8,3,3,-52]
pares = []
maior = lista[0]
menor = lista[0]
first_num = lista[0]
rep_num = 0
soma = 0
soma_negativos = 0

for i in lista:
  soma += i
  if i > maior:
    maior = i
  if i < menor:
    menor = i
  if i % 2 == 0:
    pares.append(i)
  if first_num == i:
    rep_num += 1
  if i < 0:
    soma_negativos += i

print(f"Maior: {maior:}\nMenor: {menor:}\nPares: {pares:}\nOcorrencias do primeiro numero {first_num:}: {rep_num:}\nMédia: {soma/len(lista):.2f}\nSoma negativos: {soma_negativos:}")
```

```python
class Dados:
  def __init__(self,nome,idade,cidade) -> None:
    self.nome = nome
    self.idade = idade
    self.cidade = cidade
  
  def mostrar_dados(self):
    print(f"nome: {self.nome:}\nidade: {self.idade:}\ncidade: {self.cidade:}")

lista_de_pessoas = []

while True:
  nome = input("Digite seu nome: ")
  idade = int(input("Digite sua idade: "))
  cidade = input("Digite a sua cidade: ")
  pessoa = Dados(nome,idade,cidade)
  lista_de_pessoas.append(pessoa)

  r = input("deseja continuar: ")
  if r == 'nao':
    break

for p in lista_de_pessoas:
  p.mostrar_dados()
  print("- = " * 10)
```

```python
class Carro:
  def __init__(self,modelo,km_litro) -> None:
    self.modelo = modelo
    self.km_litro = km_litro

  def get_modelo(self):
    return self.modelo

  def consumo(self):
    return (1000/self.km_litro) * 2.25
  
  def mostrar_dados(self):
    print(f"Modelo: {self.modelo:} - Km por Litro: {self.km_litro:} - consumo: R$ {self.consumo():.2f}")

carros = []
consumos = []

for i in range(0,5):
  modelo = input("Modelo do carro: ")
  km_litro = float(input("Quantos Km/L: "))
  c = Carro(modelo,km_litro)
  consumos.append(c.consumo())
  carros.append(c)

print("Relatorio Final")

mais_economico = consumos[0]
nome_carro = ''

for c in carros:
  c.mostrar_dados()
  teste = c.consumo()
  if teste < mais_economico:
    mais_economico = teste
    nome_carro = c.get_modelo()

print(f"O menor consumo é {nome_carro:}.")

```
