```python
from skimage.io import imread
from skimage.filters import roberts,prewitt,scharr
from skimage.filters import threshold_otsu
from skimage.color import rgb2gray
import matplotlib.pyplot as plt

imagem = imread('picos.jpg')

cinza = rgb2gray(imagem)
limiar = threshold_otsu(cinza)
preto_branco = cinza < limiar


bordas_roberts = roberts(preto_branco)

bordas_prewitt = prewitt(preto_branco)

bordas_scharr = scharr(preto_branco)

plt.figure(figsize=(10, 10))

plt.subplot(2, 3, 1)
plt.title('bordas_roberts')
plt.imshow(bordas_roberts, cmap='gray')

plt.subplot(2, 3, 2)
plt.title('bordas_prewitt')
plt.imshow(bordas_prewitt, cmap='gray')

plt.subplot(2, 3, 3)
plt.title('bordas_scharr')
plt.imshow(bordas_scharr, cmap='gray')

plt.show()
```
